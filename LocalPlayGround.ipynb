```python
%load_ext autoreload
%autoreload 2
```

```python
from tf.app import use
```

```python
# load the app and data from local disc
MPTF = use("app:./app", hoist=globals())
```

```python
testQuery2 = """
sentence
   sub
      word text=ανθρωπος
"""
# Execute the query
testResults2 = MPTF.search(testQuery2,silent=False)
```

```python
# Set default view in a way to limit noise as much as possible.
MPTF.displaySetup(condensed=True, multiFeatures=False, queryFeatures=False)
```

```python
MPTF.show(testResults2,start=0,end=1)
```
